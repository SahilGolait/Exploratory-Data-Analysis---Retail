### Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

### Chart 1

## Heatmap for Correlation

```python
# Construct Heatmap

corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Greens')
```



*   Sales and Profit are Moderately Correlated.

*   Discount and Profit are Negatively Correlated
*   Quantity and Profit are less Moderately Correlated.






### Chart 2

```python
# Relation between sales and profit

plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True)
plt.xlim([-200,1200])
plt.legend()
```

Profit is more than that of sale but there are some areas where profit could be increased.

### Chart 4

```python
# Create pie Chart for 'Category'

plt.figure(figsize=(15,12))
df['Sub-Category'].value_counts().plot.pie(autopct='%.0f%%')
plt.show()

# Create Count plot for 'Category'
plt.figure(figsize=(20,8))
sns.countplot(df['Sub-Category'])
plt.title('Sub-Category',fontsize=20)
```



*   From pie chart and count plot we can observed that Maximum Sold are from Binders, Paper, furnishings, Phones, storage, art, accessories sub category.
*   Minimum sold are from copiers, machines, suppliers sub catrgory.



### Chart 5

```python
# Create Count plot for 'State'

plt.figure(figsize=(15,8))
sns.countplot(x='State',data=df,order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

 

*   From above count plot it is observed that Maximum number of buyers are from California, New York and Texas. 




### Chart 6

```python
df['Quantity'].value_counts()
```

```python
# create count plot for 'Quantity'

plt.figure(figsize=(12,5))
sns.countplot(df['Quantity'])
plt.title('Quantity',fontsize=20)

```



*   2 and 3 Quantity are mostly buy by the customers




### Chart 7

```python
df['Discount'].value_counts()
```

```python
# create count plot for 'Discount'

plt.figure(figsize=(12,5))
sns.countplot(df['Discount'])
plt.xticks(rotation=90)
plt.title('Discount',fontsize=20)
```



*   Most of the products do not have any discount where some products have 20% discounts and very less products have maximum discounts.




### Chart 8

```python
# Create bar plot for 'sales' and 'profit'

ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps.plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit / loss & Sales across states',fontsize=20)
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```



*   High profit is for california, new york.
*   Maximum loss is for texas, pennsylvania, Ohio.



### Chart 9

```python
# Create 'Profit' vs 'Quantity' graph

df.groupby('Quantity')[['Profit']].sum().plot.bar(color=['blue'],figsize=(8,5))
plt.ylabel('Profit')
plt.title('Profit vs Quantity',fontsize=20)
plt.show()
```



*   Profit is maximum for those product which are taken by the customer in the group of  2, 3, 4, 5 and 7 Quantity.



### Chart 10

```python
# Create a bar plot for 'segment' and 'profit' column

df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```



*   From the above barplot 'Profit' and 'sales' are maximum in consumer segment and minimum in Home Office segment



### Chart 11

```python
# plot a barplot for 'Region' and 'Sales' column

plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region',fontsize=15)
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.show()
```



*   Segment wise sales are almost same in every region



### Chart 12

```python
# Create a barplot for 'Region'.

df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```



*   'Profit' and 'sales' are maximum in west region and minimum in south region.



### Chart 13

```python
# Create a barplot for Category column.

df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```



*   Technology and Office Supplies have high profit.
*   Furniture have less profit




### Chart 14

```python
# Create a bar plot for Sub-category.

ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps.plot.bar(color=['red','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```



*   Phones sub-category have high sales.

*   chairs have high sales but less profit compared to phones

*   Tables and Bookmarks sub-categories having huge loss

*   Copies sub-category have less sales but more profit comare to other sub-category
